## Solución sin algoritmo genético

Se establece las permutaciones de todos los puntos de un grafo

```python
def insert_multiple(x, lst):
    return [lst[:i] + [x] + lst[i:] for i in range(len(lst) + 1)]


def barter(c):
    if len(c) == 0:
        return [[]]
    return sum([insert_multiple(c[0], s) for s in barter(c[1:])], [])
```

### Buscar mínima distancia entre el conjunto de permutaciones

 - Iteración entre cada ruta
 - Se valida que la ruta es válida (se elimina en este caso)
 - Se va regunta por el peso mínimo

```python
def search_min(all_routes):
    routes_valid = []
    min_route = []
    min_weight = np.Inf

    for i in range(len(all_routes)):
        route = all_routes[i]
        is_route_valid = True
        total_weight = 0

        for j in range(len(route)):
            if j == 0:
                continue
            w = graph.get_weight(route[j-1], route[j])
            total_weight = total_weight + w

        w = graph.get_weight(route[-1], route[0])
        total_weight = total_weight + w

        all_routes[i].append(route[0])
        routes_valid.append(all_routes[i])

        if total_weight < min_weight:
            min_weight = total_weight
            min_route = all_routes[i]

    return {
        "routes_valid": routes_valid,
        "min_route": min_route,
        "min_weight": min_weight
    }
```

Se pone a prueba el algoritmo.

 - Se crea el grafo
 - Se calculan todas las rutas
 - Se pasa para el algoritmo

```python
graph = Graph.creategraph(5, 1.0, directed=False)
cities = graph.nodes
cities_bartered = barter(cities)
result = search_min(cities_bartered)

print("Total: ", len(result["routes_valid"]))
print("Route min: ", result["min_route"])
print("Weight min: ", result["min_weight"])
graph.draw()
```

```python
for num_nodes in range(4, 8):
    graph = Graph.creategraph(num_nodes, pro_edges=1.0, directed=False)
    t = time()
    cities_bartered = barter(graph.nodes)
    result = search_min(cities_bartered)
    lapsed = time() - t
    print("Total: ", len(result["routes_valid"]))
    print("Weight min: ", result["min_weight"])
    print('Time: ', lapsed)
    graph.draw()
```

# Algoritmo Génetico

El enfoque

Comencemos con algunas definiciones, reformuladas en el contexto del TSP:

     - Gen: Una ciudad (representada como un punto en el grafo)
     - Individual: (también conocido como "cromosoma"): una sola ruta que satisface las condiciones anteriores
     - Population: Colección de rutas posibles (es decir, colección de individuos)
     - Parents: Dos rutas que se combinan para crear una nueva ruta.
     - Mating pool: Una colección de padres que se utilizan para crear nuestra siguiente población (creando así la próxima generación de rutas)
     - Fitness: una función que nos dice qué tan buena es cada ruta (en nuestro caso, cuán corta es la distancia)
     - Mutation: una forma de introducir la variación en nuestra población mediante el intercambio aleatorio de dos ciudades en una ruta
     - Elitism: una forma de llevar a los mejores individuos a la siguiente generación.

1. Create the population

2. Determine fitness

3. Select the mating pool

4. Breed

5. Mutate

6. Repeat

```python
from graph.Graph import Graph
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import operator
import random
```

```python
graph = Graph.creategraph(20, 1.0, directed=False)
cities = graph.nodes
```

Se define la clase City

```python
class City:
    def __init__(self, id):
        self.id = id

    def distance(self, id):
        return graph.get_weight(self.id, id)
```

Trataremos la condición física como la inversa de la distancia de la ruta. Queremos minimizar la distancia de la ruta, por lo que un mayor puntaje de fitness es mejor

```python
class Fitness:
    def __init__(self, route):
        self.route = route
        self.distance = 0
        self.fitness = 0.0

    def route_distance(self):
        if self.distance == 0:
            path_distance = 0
            for i in range(0, len(self.route)):
                from_city = self.route[i]
                to_city = self.route[i + 1] if i + 1 < len(self.route) else self.route[0]

                path_distance += from_city.distance(to_city.id)
            self.distance = path_distance
        return self.distance
    
    def route_fitness(self):
        if self.fitness == 0:
            self.fitness = 1 / float(self.route_distance())
        return self.fitness
```

Ahora podemos hacer que nuestra población inicial (también conocida como primera generación). 

PSe crea una función que produzca rutas que satisfagan nuestras condiciones 

Para crear un individuo, seleccionamos al azar el orden en el que visitamos cada ciudad:

```python
def create_route(city_list):
    route = random.sample(city_list, len(city_list))
    return route


def initial_population(pop_size, city_list):
    population = []

    for _ in range(0, pop_size):
        population.append(create_route(city_list))
    return population
```

Para simular nuestra "supervivencia del más apto", podemos utilizar Fitness para clasificar a cada individuo en la población. 

Nuestra salida será una lista ordenada con los ID de ruta y cada puntaje de condición física asociado.

```python
def rank_routes(population):
    fitness_results = {}
    for i in range(0, len(population)):
        fitness_results[i] = Fitness(population[i]).route_fitness()
    return sorted(fitness_results.items(), key=operator.itemgetter(1), reverse=True)
```

Seleccione la mating pool

Hay algunas opciones para seleccionar los padres que se utilizarán para crear la próxima generación.

Los enfoques más comunes son la selección proporcional al fitness (también conocida como "selección de la rueda de la ruleta") o la selección de torneos:

     - Selección proporcional de aptitud (la versión implementada a continuación): La aptitud de cada individuo en relación con la población se utiliza para asignar una probabilidad de selección. Piense en esto como la probabilidad ponderada de aptitud de ser seleccionado.
     
     - Selección de torneo: un número determinado de individuos se seleccionan al azar de la población y el que tiene el mayor nivel de aptitud en el grupo se elige como el primer padre. Esto se repite para elegir al segundo padre.
     
 Otra característica del diseño a considerar es el uso del elitismo.
 
 Con el elitismo, los individuos con mejor desempeño de la población se trasladarán automáticamente a la siguiente generación, asegurando que los individuos con más éxito persistan.

```python
def selection(pop_ranked, elite_size):
    selection_results = []
    df = pd.DataFrame(np.array(pop_ranked), columns=["Index", "Fitness"])
    df['cum_sum'] = df.Fitness.cumsum()
    df['cum_perc'] = 100 * df.cum_sum / df.Fitness.sum()

    for i in range(0, elite_size):
        selection_results.append(pop_ranked[i][0])
    for _ in range(0, len(pop_ranked) - elite_size):
        pick = 100 * random.random()
        for i in range(0, len(pop_ranked)):
            if pick <= df.iat[i, 3]:
                selection_results.append(pop_ranked[i][0])
                break
    return selection_results
```

Ahora que tenemos los ID de las rutas que conformarán nuestro grupo de apareamiento desde la función de selección, podemos crear el grupo de apareamiento.

Simplemente estamos extrayendo a los individuos seleccionados de nuestra población.

```python
def mating_pool(population, selection_results):
    mating_pool_array = []
    for i in range(0, len(selection_results)):
        index = selection_results[i]
        mating_pool_array.append(population[index])
    return mating_pool_array
```

### Raza

Con la creación de nuestro grupo de apareamiento, podemos crear la próxima generación en un proceso denominado cruce.

El TSP es único ya que necesitamos incluir todas las ubicaciones exactamente una vez.

Podemos utilizar una función de reproducción especial denominada **cruce ordenado**.

En el cruce ordenado, seleccionamos al azar un subconjunto de la primera cadena principal y luego completamos el resto de la ruta con los genes del segundo padre en el orden en que aparecen, sin duplicar ningún gen, en el subconjunto seleccionado del primer padre.

```python
def breed(parent1, parent2):
    child_p1 = []

    gene_a = int(random.random() * len(parent1))
    gene_b = int(random.random() * len(parent1))

    for i in range(min(gene_a, gene_b), max(gene_a, gene_b)):
        child_p1.append(parent1[i])

    child_p2 = [item for item in parent2 if item not in child_p1]

    child = child_p1 + child_p2
    return child

def breed_population(mating_pool_array, elite_size):
    children = []
    length = len(mating_pool_array) - elite_size
    pool = random.sample(mating_pool_array, len(mating_pool_array))

    for i in range(0, elite_size):
        children.append(mating_pool_array[i])

    for i in range(0, length):
        child = breed(pool[i], pool[len(mating_pool_array) - i - 1])
        children.append(child)

    return children
```

## Mutación

La mutación cumple una función importante en GA, ya que ayuda a evitar la convergencia local al introducir nuevas rutas que nos permitirán explorar otras partes del espacio de la solución.

Similar al cruce, el TSP tiene una consideración especial cuando se trata de la mutación.

No podemos abandonar las ciudades. En su lugar, usaremos mutación de intercambio. Esto significa que, con baja probabilidad especificada, dos ciudades intercambiarán lugares en nuestra ruta. Haremos esto para un individuo en nuestra función de mutar:

```python
def mutate(individual, mutation_rate):
    for swapped in range(len(individual)):
        if random.random() < mutation_rate:
            swap_with = int(random.random() * len(individual))

            city1 = individual[swapped]
            city2 = individual[swap_with]

            individual[swapped] = city2
            individual[swap_with] = city1

    return individual

def mutate_population(population, mutation_rate):
    mutated_pop = []

    for ind in range(0, len(population)):
        mutated_ind = mutate(population[ind], mutation_rate)
        mutated_pop.append(mutated_ind)

    return mutated_pop
```

## Generación

Juntemos estas piezas para crear una función que produzca una nueva generación.

Primero, clasificamos las rutas en la generación actual usando rankRoutes.

Luego, determinamos nuestros padres potenciales ejecutando la función de selección, que nos permite crear el grupo de apareamiento utilizando la función matingPool.

Finalmente, creamos nuestra nueva generación utilizando la función breedPopulation.

Y luego aplicando la mutación utilizando la función mutatePopulation.

```python
def nextGeneration(current_gen, elite_size, mutation_rate):
    pop_ranked = rank_routes(current_gen)
    selection_results = selection(pop_ranked, elite_size)
    mating_pool_array = mating_pool(current_gen, selection_results)
    children = breed_population(mating_pool_array, elite_size)
    next_generation = mutate_population(children, mutation_rate)

    return next_generation
```

## Evolución en movimiento

Todo lo que necesitamos hacer es crear la población inicial,

y luego podemos recorrer tantas generaciones como deseemos. Por supuesto, también queremos ver la mejor ruta y cuánto hemos mejorado, así que capturamos las capturamos.

Presentamos los resultados

```python
def geneticAlgorithm(population, pop_size, elite_size, mutation_rate, generations):
    pop = initial_population(pop_size, population)
    progress = []
    progress.append(1 / rank_routes(pop)[0][1])

    print("Initial distance: " + str(1 / rank_routes(pop)[0][1]))

    for _ in range(0, generations):
        pop = nextGeneration(pop, elite_size, mutation_rate)
        progress.append(1 / rank_routes(pop)[0][1])

    print("Final distance: " + str(1 / rank_routes(pop)[0][1]))
    best_route_index = rank_routes(pop)[0][0]
    best_route = pop[best_route_index]

    plt.plot(progress)
    plt.ylabel('Distance')
    plt.xlabel('Generation')
    plt.show()

    return best_route
```

```python
cityList = []
for gr_city in cities:
    cityList.append(City(id=gr_city))


geneticAlgorithm(
    population=cityList,
    pop_size=100,
    elite_size=20,
    mutation_rate=0.01,
    generations=500
)
```
